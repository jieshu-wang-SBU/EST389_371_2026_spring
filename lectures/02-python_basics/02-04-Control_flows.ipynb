# Python Basics (4): control flows
## Conditional Statements: `if`, `elif`, `else`

Conditional statements allow your program to make decisions based on conditions.
- Conditions are checked top to bottom
- Only one branch runs


```python
number = 15

if number >= 100:
    print("Larger than 100")
elif number < 0:
    print("Negative number")
else:
    print("Positive number smaller than 100")
```

## `for` Loop (Iteration over a Sequence)
A for loop iterates over elements of an *iterable* (list, string, range, etc.).
Common pattern:

`for item in iterable:`

`...`

```python
colors = ["red", "green", "blue"]

for color in colors:
    print(color)
```

## `while` Loop (Repeat Until Condition Fails)
A while loop repeats as long as a condition is True.

```python
number = 0

while number < 3:
    print(number)
    number += 1
```

### `while` Loop with `else`
The `else` block runs only if the loop finishes normally (no break).

`else` runs if the loop was not interrupted.”

```python
number = 0

while number < 3:
    print(number)
    number += 1
else:
    print(f"End with {number}")
```

## `break`: Exit the Loop Early
`break` in a `for` loop


```python
numbers = range(0, 10)
print(numbers)

for number in numbers:
    if number == 5:
        break # break immediately stops the loop.
    print(number)
```

### `break` in a `while` loop


```python
count_num = 0

while count_num < 5:
    if count_num == 3:
        break
    print(count_num)
    count_num += 1
```

## `continue`: Skip Current Iteration
`continue` skips the rest of the loop body and moves to the next iteration.

```python
numbers = range(0, 5)

for number in numbers:
    if number == 2:
        continue
    print(number)
```
