# Visualize proportions
## Multiple barplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "company": [
        "A", "B", "C", "D", "E",
        "A", "B", "C", "D", "E",
        "A", "B", "C", "D", "E"
    ],
    "year": [
        2022, 2022, 2022, 2022, 2022,
        2023, 2023, 2023, 2023, 2023,
        2024, 2024, 2024, 2024, 2024
    ],
    "market_share": [
        30, 25, 20, 15, 10,   # 2022
        28, 27, 22, 13, 10,   # 2023
        26, 29, 24, 12, 9     # 2024
    ]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
# plot side-by-side bar chart

fig, ax=plt.subplots(nrows=1, ncols=1, figsize=(6.3,4), tight_layout=True, sharex=True, sharey=True)

sns.barplot(data=df, x="year", y="market_share", hue='company', ax=ax)
```

## Stacked bar charts (proportion) over time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

import pandas as pd
import numpy as np

np.random.seed(42)

years = list(range(2005, 2025))

data = []

for i, year in enumerate(years):
    women_pct = 35 + i * 1.2 + np.random.normal(0, 1)
    women_pct = min(max(women_pct, 0), 100)
    men_pct = 100 - women_pct

    data.append([year, "women", round(women_pct, 1)])
    data.append([year, "men", round(men_pct, 1)])

df = pd.DataFrame(data, columns=["year", "gender", "percentage"])

df.head()
```

```python

pivot_df = df.pivot(index="year", columns="gender", values="percentage")

g=pivot_df.plot.bar(figsize=(10,4), stacked=True, width=0.9)
plt.title("stacked participation share")
plt.ylabel("percentage (%)")
```
