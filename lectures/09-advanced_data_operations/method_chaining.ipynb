# Method Chaining

_Method chaining_ is a coding style where multiple DataFrame or Series operations are invoked sequentially using the dot (.) operator, creating a single, fluid data transformation pipeline. This approach improves readability and maintainability by eliminating the need for intermediate variables and presenting the data flow as a clear sequence of steps.

## Example: Comparing code with and without method chaining

goal: Analyze survival rate and average fare per person by family size.
We will first create two new columns:

- family_size
- fare_per_person

Then compute summary statistics.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../06-Exploratory_data_analysis/data/Titanic-Dataset.csv")
df.head()
```

### Traditional (Non-Chained) Approach:

```python
# Step 1: create family size
df["family_size"] = df["SibSp"] + df["Parch"] + 1

# Step 2: create fare per person
df["fare_per_person"] = df["Fare"] / df["family_size"]

# Step 3: group and aggregate
result = df.groupby("family_size").agg(
    survival_rate = ("Survived", "mean"),
    avg_fare_per_person = ("fare_per_person", "mean"),
    passenger_count = ("PassengerId", "count")
)

# Step 4: sort results
result = result.sort_index()

result
```

Issues with this approach
- many intermediate mutations
- dataframe gets modified multiple times
- harder to track workflow in longer pipelines

### Method chaining version:



```python
result = (
    df
    .assign(
        family_size = lambda x: x["SibSp"] + x["Parch"] + 1,
        fare_per_person = lambda x: x["Fare"] / x["family_size"]
    )
    .groupby("family_size")
    .agg(
        survival_rate = ("Survived", "mean"),
        avg_fare_per_person = ("fare_per_person", "mean"),
        passenger_count = ("PassengerId", "count")
    )
    .sort_index()
)

result
```
