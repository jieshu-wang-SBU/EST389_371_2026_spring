# Dataclass
Dataclass in Python is very useful. It is designed to simplify the creation of classes that primarily store data by automatically generating boilerplate methods.

Dataclasses are ideal for data transfer objects (DTOs), configuration containers, or any situation where a class's primary purpose is to hold structured data. They are generally preferred over dictionaries or tuples when you want type safety, default values, and clear attribute names with dot-notation access

```python
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    age: int
    major: str

# Create instance
s1 = Student(name="Alice", age=21, major="Data Science")

print(s1)
```

```python
print(f"The student name is {s1.name}, and age is {s1.age} with major {s1.major}")
```

In file path configuration

```python
@dataclass
class NewsOutletConfigVars:
    news_outlet_name: str
    scraped_news_file: str
    nyt_topic_extraction_path: str
    news_category_path: str
    geocoding_path: str

# We can set up the configurations:
washington_post = NewsOutletConfigVars(news_outlet_name="Washington Post", scraped_news_file=".../data/washington_post.csv", nyt_topic_extraction_path=".../data/topic_extraction.csv", news_category_path=".../data/news_category.csv", geocoding_path="...")

print(washington_post.news_category_path)
```
