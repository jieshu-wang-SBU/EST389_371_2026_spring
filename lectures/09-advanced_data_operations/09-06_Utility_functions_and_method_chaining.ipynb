## z-score util function

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# read the data
nls97 = pd.read_csv("../05-data_cleaning/data/nls97.csv")

nls97
```

If we are tasked to compute the z-scores of `wageincome`, `satverbal`, `satmath`, `gpaoverall`, `gpaenglish`, `gpamath`, `gpascience`, at the same time, all the z-scores for those variables based on their `maritalstatus`, we need to compute many many times.

For example:

```python
nls97['wageincome_z_score'] = (nls97['wageincome'] - nls97['wageincome'].mean())/nls97['wageincome'].std()
nls97['satverbal_z_score'] = (nls97['satverbal'] - nls97['satverbal'].mean())/nls97['satverbal'].std()
nls97['satmath_z_score'] = (nls97['satmath'] - nls97['satmath'].mean())/nls97['satmath'].std()
nls97['gpaoverall_z_score'] = (nls97['gpaoverall'] - nls97['gpaoverall'].mean())/nls97['gpaoverall'].std()
nls97['gpaenglish_z_score'] = (nls97['gpaenglish'] - nls97['gpaenglish'].mean())/nls97['wageincome'].std()
nls97['gpascience_z_score'] = (nls97['gpascience'] - nls97['gpascience'].mean())/nls97['gpascience'].std()
```

The above cell is not only very long, every time you need to compute a new z_score, you need to write it again. Alternative way is to write a function.

```python
def z_score(series:pd.Series)->pd.Series:
    return (series - series.mean()) / series.std()

for col in ['wageincome', 'satverbal', 'satmath', 'gpaoverall', 'gpaenglish', 'gpamath', 'gpascience']:
    nls97[f"{col}_z_score"] = z_score(nls97[col])

nls97[[i for i in nls97.columns if 'z_score' in i]]
```

You can call it in groupby as well

```python
for col in ['wageincome', 'satverbal', 'satmath', 'gpaoverall', 'gpaenglish', 'gpamath', 'gpascience']:
    nls97[f"{col}_z_score_by_marital_status"] = nls97.groupby("maritalstatus")[col].transform(z_score)

nls97[[i for i in nls97.columns if 'z_score_by' in i]]
```

# Method Chaining

_Method chaining_ is a coding style where multiple DataFrame or Series operations are invoked sequentially using the dot (.) operator, creating a single, fluid data transformation pipeline. This approach improves readability and maintainability by eliminating the need for intermediate variables and presenting the data flow as a clear sequence of steps.

## Example: Comparing code with and without method chaining

goal: Analyze survival rate and average fare per person by family size.
We will first create two new columns:

- family_size
- fare_per_person

Then compute summary statistics.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../06-Exploratory_data_analysis/data/Titanic-Dataset.csv")
df.head()
```

### Traditional (Non-Chained) Approach:

```python
# Step 1: create family size
df["family_size"] = df["SibSp"] + df["Parch"] + 1

# Step 2: create fare per person
df["fare_per_person"] = df["Fare"] / df["family_size"]

# Step 3: group and aggregate
result = df.groupby("family_size").agg(
    survival_rate = ("Survived", "mean"),
    avg_fare_per_person = ("fare_per_person", "mean"),
    passenger_count = ("PassengerId", "count")
)

# Step 4: sort results
result = result.sort_index()

result
```

Issues with this approach
- many intermediate mutations
- dataframe gets modified multiple times
- harder to track workflow in longer pipelines

### Method chaining version:



```python
result = (
    df
    .assign(
        family_size = lambda x: x["SibSp"] + x["Parch"] + 1,
        fare_per_person = lambda x: x["Fare"] / x["family_size"]
    )
    .groupby("family_size")
    .agg(
        survival_rate = ("Survived", "mean"),
        avg_fare_per_person = ("fare_per_person", "mean"),
        passenger_count = ("PassengerId", "count")
    )
    .sort_index()
)

result
```
