# Groupby


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("./data/IMDB Top 250 Movies.csv")

# let's clean the data type
numeric_columns = ["budget",'box_office']

for col in numeric_columns:
    df[col] = df[col].apply(pd.to_numeric, errors='coerce')

df
```

```python
df.info()
```

## Basic group by

Average rating by genre

```python
genre_rating = df.groupby("genre")["rating"].count()

genre_rating.sort_values(ascending=False)
```

Actually, we can see that one movie might have multiple genres. If we wanted to group by individual genre, we could do `.explode` and then `.groupby`

```python
df_exploded_by_genre = df.assign(
    genre_list = lambda df: df["genre"].str.split(",")
).explode('genre_list')

df_exploded_by_genre.head()
```

```python
# compute individual genre's average rating

genre_rating = df_exploded_by_genre.groupby(by='genre_list')['rating'].mean().reset_index().rename(columns={"genre_list": "genre", "rating":"mean_rating"})

genre_rating.sort_values(by='mean_rating',ascending=False)
```

## Count movies by Genre or Year

```python
genre_count = df_exploded_by_genre.groupby("genre_list")["name"].count().reset_index().rename(columns={"name":"movie_count"})

genre_count.sort_values(by="movie_count", ascending=False)
```

```python
yearly_count = df.groupby("year")["name"].count().reset_index().rename(columns={"name":"movie_count"})

yearly_count
```

```python
sns.lineplot(data=yearly_count, x="year", y="movie_count")
```

## Group by multiple columns

```python
yearly_movies_by_genre = df_exploded_by_genre.groupby(['year', 'genre_list'], as_index=False)['name'].count().rename(columns={"name":"movie_count","genre_list":"genre"})

yearly_movies_by_genre
```

## Multiple aggregations

### Use a dictionary

- One groupby with multiple metrics
- Output becomes multi-level columns

```python
genre_summary = df_exploded_by_genre.groupby("genre_list").agg(
    {"rating": ["mean", "max"],
    "box_office": "mean"}
)

genre_summary
```

```python
# It generates multi-level index
genre_summary.columns
```

```python
# I prefer converting them to one-level index
genre_summary.columns = ["_".join(i) for i in genre_summary.columns]
genre_summary
```

```python
genre_summary.columns
```

### Custom naming
#### Custom naming with tuples

```python
genre_summary = df_exploded_by_genre.groupby("genre_list").agg(
    rating_mean=("rating", 'mean'),
    rating_max=("rating", "max"),
    box_office_mean=("box_office", lambda x: x.mean()+1) # you can also use custom functions
).rename(columns={"genre_list":"genre"})

genre_summary
```

#### Custom naming with `pd.NamedAgg()`
This is a very powerful method that enables highly customizable aggregation functions

```python
genre_summary = df_exploded_by_genre.groupby("genre_list", as_index=False).agg(
    rating_mean = ('rating', 'mean'),
    rating_max = pd.NamedAgg(column='rating', aggfunc='max'),
    box_office = pd.NamedAgg(column='box_office', aggfunc=lambda x: x.mean()),
    average_run_time = pd.NamedAgg(column='run_time', aggfunc=lambda x: pd.to_timedelta(x, unit='m', errors='coerce').mean().total_seconds()/60)
).rename(columns={"genre_list":"genre"})

genre_summary
```

## Groupby and Transform

### Example 1: compute the proportion of box office of each movie in each genre

```python
df_exploded_by_genre['box_office_pct_in_genre'] = df_exploded_by_genre.groupby("genre_list")['box_office'].transform(lambda x: x*100/x.sum())

df_exploded_by_genre[['name','genre_list','box_office','box_office_pct_in_genre']]
```

We can further compute the diversity or uncertainty index (HHI or Shannon Entropy) in each genre in terms of box office, using `.groupby()`

_Herfindahl-Hirschman index (HHI)_ is a measure of market concentration calculated by summing the squares of the percentage market shares of all firms in an industry.

Variations of HHI:
- Basic HHI (or Simpson Index): $\lambda = \sum_{i=1}^{n} s_i^2$
- Inverse HHI (or inverse Simpson index): $\frac{1}{\lambda} = \frac{1}{\sum_{i=1}^{n} s_i^2}$
- Gini-Simpson Index (or Gini's diversity index in machine learning): $1-\lambda = 1- \sum_{i=1}^{n} s_i^2$
    - The higher it is, the more diverse the system is




```python
genre_diversity = df_exploded_by_genre.groupby("genre_list", as_index=False).agg(
    box_office_gini_diversity = pd.NamedAgg(column='box_office_pct_in_genre', aggfunc=lambda x: 1-((x/100)**2).sum()),
)

genre_diversity
```

### Example 2: Compute z-score of ratings

A Z-score (standard score) measures how many standard deviations a data point (x) is above or below the population mean (μ).

$z=\frac{x-\mu}{\sigma}$

where $\sigma$ represents the standard deviation.


```python

df_exploded_by_genre['rating_z_score'] = df_exploded_by_genre.groupby("genre_list")['rating'].transform(lambda x: (x-x.mean())/x.std())

df_exploded_by_genre[['name','genre_list','rating','rating_z_score']]
```
